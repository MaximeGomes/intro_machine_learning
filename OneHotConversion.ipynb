*Convert categorical variables to numeric values through One-Hot encoding*

```python
import pandas as pd
```

```python
# We will be using the FMA dataset (from KPrototypes) for this exercise

#Load Excel file into a dataframe
FMADf = pd.read_excel("all_datasets/FreeMusicArchive.xlsx", sheet_name="FMADataset")

# Exclude track_id from clustering algorithm
FMADf = FMADf.loc[:, FMADf.columns != 'track_id']

# Convert categorical variables to string datatype
FMADf['artist_name'] = FMADf['artist_name'].astype(str)
FMADf['type'] = FMADf['type'].astype(str)
FMADf['genre_top'] = FMADf['genre_top'].astype(str)

#Display info 
```

```python
# Processing categorical variables
categorical = FMADf.select_dtypes(include='object')
categorical = pd.get_dummies(categorical)
```
